```python
def calculate_tax(item_price,tax_rate,discount):
    tax = item_price * (tax_rate/100)
    total = item_price+tax - discount
    return total,tax
```

```python
item_price = 120.00
tax_rate = 20.00
discount = 10.00
total,tax = calculate_tax(item_price,tax_rate,discount)
print(f"Total: {total}")
print(f"tax applied :{tax}")
```
